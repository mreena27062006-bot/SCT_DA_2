```python
#importing pandas library
import pandas as pd
print("Pandas loaded successfully!")
```

```python
#loading the data 
df = pd.read_csv('superstore.csv', encoding='latin1')
print("Dataset loaded!")

# checking for totaal rows and columns
print("Shape:", df.shape)
```

```python
# displaying 5 rows from the dataset
df.head()
```

```python
# To See Column Info
df.info()
```

```python
#Statistical Summary
df.describe()
```

```python
# See all current column names first
print(df.columns.tolist())
```

```python
# Renaming that corrupted column
df.columns = [col.encode('ascii', 'ignore').decode('ascii').strip('') for col in df.columns]

# Check columns again
print(df.columns.tolist())
```

```python
df.describe()
```

```python
# Rename the empty column to 'Returns'
df.rename(columns={'': 'Returns'}, inplace=True)

# Verify it's fixed
print(df.columns.tolist())
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking for missing values 
print("Missing values per column:")
print(df.isnull().sum())
print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
# Remove rows where EVERY cell is empty
df = df.dropna(how='all')

# Fill remaining empty cells with 'Unknown'
df = df.fillna('Unknown')

print("Missing values after cleaning:", df.isnull().sum().sum())
```

```python
#Check and Remove Duplicates
print("Duplicate rows before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicate rows after:", df.duplicated().sum())
print("Total rows remaining:", len(df))
```

```python
print(df.dtypes)
```

```python
# Convert Dates to Proper Format
df['Order.Date'] = pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

print("After conversion:")
print(df[['Order.Date', 'Ship.Date']].dtypes)
```

```python
#Extract Year, Month from Dates
df['Order.Year'] = df['Order.Date'].dt.year
df['Order.Month'] = df['Order.Date'].dt.month
df['Order.Month_Name'] = df['Order.Date'].dt.month_name()

print(df[['Order.Date', 'Order.Year', 'Order.Month', 'Order.Month_Name']].head())
```

```python
# Fix  Profit  and Sales as Numbers
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

print("Profit type:", df['Profit'].dtype)
print("Sales type:", df['Sales'].dtype)
```

```python
print("=== FINAL DATASET SUMMARY ===")
print("Total Rows:", len(df))
print("Total Columns:", len(df.columns))
print("Missing Values:", df.isnull().sum().sum())
print("Duplicate Rows:", df.duplicated().sum())
print("\nAll Column Names:")
print(df.columns.tolist())
```
